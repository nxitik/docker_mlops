```python
import pandas as pd
df = pd.read_csv("data_extract_from_s3/healthcare_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Name", "Doctor", "Hospital", "Room Number"], axis = 1, inplace = True)
df.head()
```

```python
df.drop(["Insurance Provider", "Date of Admission","Discharge Date"], axis = 1, inplace = True)
```

```python
df['Billing Amount'] = df['Billing Amount'].round(-3)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = []
for col in df.columns:
    if df[col].dtypes == "object":
        cat_cols.append(col)
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rfc.predict(X_test).round()
y_pred
print("Validation Accuracy", accuracy_score(y_test, y_pred))
```
